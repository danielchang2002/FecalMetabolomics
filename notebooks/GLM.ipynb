## Setup

```r
df = read.csv(file = '../data/cleaned.csv')
df
```

## GLM

```r
library(lme4)
```

#### For each biochemical, create a mixed effects glm. Timepoint is dependent variable, mouseID is a random effect and gender and abundance are fixed effect

```r
biochemicals = colnames(df)[2:824]
tail(biochemicals)
```

#### Initialize p_val dataframe

```r
p_vals = data.frame(
    biochemical = biochemicals,
    p_val = -1
)
head(p_vals)
```

#### Model Function

```r
get_p_val = function(data) {
    model = glmer(AfterProbiotic ~ (1|mouseID) + isMale + ., data=data, family=binomial)
    coef = coef(summary(model))[,4]
    p_val = coef[4]
    return (p_val)
}
```

```r
# Test with first biochemical
get_p_val(df[, c(1, 2, 825, 826)])
```

#### Run model on each biochemical

```r
for (biochemIDX in 2:824) {
    data = df[, c(1, biochemIDX, 825, 826)]
    p_val = tryCatch(
      get_p_val(data),
      error=function(e) {return (1)}
    )
    p_vals[biochemIDX - 1, 2] = p_val
}
```

```r
head(p_vals)
```

#### Select Biochemicals with p <= 0.05

```r
p_vals[p_vals[, 'p_val'] <= 0.05,]
```

```r
Multiple Hypothesis Correction
```

```r
p_vals_adjusted = data.frame(p_vals)
p_vals_adjusted[, 2] = p.adjust(p_vals_adjusted[, 2], method='BH')
head(p_vals_adjusted)
```
